```python
# Code Generated by Sidekick is for learning and experimentation purposes only.
import requests
from bs4 import BeautifulSoup
import pandas as pd
import re

entries = []

# --------- Penn State Fencing Roster 2017-18 (FIXED) ---------
psu_url = "https://gopsusports.com/sports/fencing/roster/season/2017-18"
psu_resp = requests.get(psu_url)
psu_soup = BeautifulSoup(psu_resp.text, "html.parser")
for a in psu_soup.find_all('a', class_='table__roster-name'):
    # Try pulling name from <span>, else directly from link
    span = a.find('span')
    if span and span.get_text(strip=True):
        name = span.get_text(strip=True)
    else:
        name = a.get_text(strip=True)
    if name and " " in name:  # Ensure plausible name format
        entries.append({'Name': name, 'School': 'Penn State'})

# --------- Vassar Men's Fencing Roster 2014-15 ---------
vassar_url = 'https://www.vassarathletics.com/sports/mens-fencing/roster/2014-15'
vassar_resp = requests.get(vassar_url)
vassar_soup = BeautifulSoup(vassar_resp.text, 'html.parser')
pattern_vassar = re.compile(r'^/sports/mens-fencing/roster/.+/\d+$')
exclude_texts_vassar = {"Full Bio", "Career Stats", "View Full Bio", ""}
for link in vassar_soup.find_all('a', href=True):
    link_text = link.text.strip()
    normalized_name = re.sub(r'\s+', ' ', link_text)
    if (
        pattern_vassar.match(link['href'])
        and normalized_name not in exclude_texts_vassar
        and len(normalized_name.split()) >= 2
    ):
        entries.append({'Name': normalized_name, 'School': 'Vassar'})

# --------- UNC Fencing Roster (2013-14) ---------
unc_url = "https://goheels.com/sports/fencing/roster/2013-14"
unc_resp = requests.get(unc_url)
unc_soup = BeautifulSoup(unc_resp.content, "html.parser")
for span in unc_soup.find_all("span", class_="text-theme-link-light s-text-regular-bold"):
    name = span.get_text(strip=True)
    if name:
        entries.append({'Name': name, 'School': 'UNC'})

# --------- UCSD Fencing Roster 2014-15 (improved) ---------
ucsd_url = "https://ucsdtritons.com/sports/fencing/roster/2014-15"
ucsd_response = requests.get(ucsd_url)
ucsd_soup = BeautifulSoup(ucsd_response.text, 'html.parser')
pattern_ucsd = re.compile(r'^/sports/fencing/roster/[\w\-]+/\d+$')
exclude_texts_ucsd = {et.lower().strip() for et in ['Full Bio', 'View Full Bio', 'Career Stats', '']}
for a in ucsd_soup.find_all('a', href=True):
    if pattern_ucsd.match(a['href']):
        name = re.sub(r'\s+', ' ', a.get_text()).strip()
        if " " in name and name.lower() not in exclude_texts_ucsd:
            entries.append({'Name': name, 'School': 'UCSD'})

# --------- UCSD Fencing Roster 2015-16 ---------
ucsd_pt_2_url = "https://ucsdtritons.com/sports/fencing/roster/2015-16"
ucsd_pt_2_resp = requests.get(ucsd_pt_2_url)
ucsd_soup_2 = BeautifulSoup(ucsd_pt_2_resp.text, 'html.parser')
pattern_ucsd_2 = re.compile(r'^/sports/fencing/roster/[\w\-]+/\d+$')
exclude_texts_ucsd_2 = {et.lower().strip() for et in ['Full Bio', 'View Full Bio', 'Career Stats', '']}
for a in ucsd_soup_2.find_all('a', href=True):
    if pattern_ucsd_2.match(a['href']):
        name = re.sub(r'\s+', ' ', a.get_text()).strip()
        if " " in name and name.lower() not in exclude_texts_ucsd_2:
            entries.append({'Name': name, 'School': 'UCSD'})

# --------- Save Combined Unique Names and Schools to CSV ---------
df = pd.DataFrame(entries).drop_duplicates(subset=['Name', 'School']).sort_values(by=['School', 'Name'])
df.to_csv("all_fencing_names.csv", index=False)
print(f"Scraped {len(df)} unique name+school pairs saved to all_fencing_names.csv")
```
